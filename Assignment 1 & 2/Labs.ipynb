```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
 
image = cv2.imread(r"1.png", 1)
half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)
rescale = cv2.resize(image, (224, 224))
stretch_near = cv2.resize(image, (780, 540),
               interpolation = cv2.INTER_LINEAR)
Titles =["Original", "Half", "Rescale", "Interpolation Nearest"]
images =[image, half, rescale, stretch_near]
count = 4
 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
 
plt.show()
```
